# Plot Groundwater Head (Cross-section)

```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import flopy
import flopy.utils as fu
import flopy.modflow as mf
```

```python
hds = fu.HeadFile('landelino-model\landelina.hds')
x = hds.get_data()
hds.plot(colorbar=True)
```

```python
def plot(layer,row):
    h = x[layer][row]
    plt.plot(h,'o', ls='-', ms=2,label = 'Layer '+ str(layer+1)+' & row '+ str(row+1))
    plt.xlabel('Column number')
    plt.ylabel('Groundwater Head (m)')
    plt.legend()
    plt.xlim(0,110)
```

```python
plot(1,60)
plot(1,40)
plot(1,50)
```
